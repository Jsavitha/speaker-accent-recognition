```python
#imbalance dataset
from imblearn.over_sampling import SMOTE
```

```python
dataset= pd.read_csv("accent-mfcc-data-1.csv")
```

```python
X = dataset.iloc[:, 1:].values
y = dataset.iloc[:, 0].values
```

```python
smote = SMOTE(random_state=42)  # You can specify random_state for reproducibility
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
import pandas as pd
from collections import Counter

print("Class distribution before SMOTE:", Counter(y))
print("Class distribution after SMOTE:", Counter(y_resampled))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert the NumPy array to a Pandas Series
y_resampled_series = pd.Series(y_resampled)

# Plot the distribution of the target variable
y_resampled_series.value_counts().plot(kind='bar', title='Count (target)')
plt.show()

```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
#splitting the dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled , y_resampled, test_size = 0.2,)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Training the Decision Tree Classification Model on the Training Set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy', max_depth=7, min_samples_leaf=3)
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.score(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#Training the Random Forest Model on the Training Set
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(min_samples_leaf=2,
                            min_samples_split=3,n_estimators=120,
                            max_features='auto',bootstrap= False,
                             max_depth =21,criterion= 'entropy')
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
